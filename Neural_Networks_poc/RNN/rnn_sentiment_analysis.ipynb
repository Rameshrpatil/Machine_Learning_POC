```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv('../Datasets/amazon.csv')
```

```python
df
```

```python
X=df['reviewText']
Y=df['Positive']
```

```python
from sklearn.model_selection import train_test_split
Xtr,Xte,Ytr,Yte=train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
from tensorflow.keras.preprocessing.text import Tokenizer
tokenizer=Tokenizer(oov_token='<nothing>')
```

```python
tokenizer.fit_on_texts(X)
```

```python
tokenizer.word_index
```

```python
tokenizer.word_counts
```

```python
tokenizer.document_count
```

```python
sequences =tokenizer.texts_to_sequences(X)
```

```python
sequences
```

```python
from tensorflow.keras.utils import pad_sequences
```

```python
sequences =pad_sequences(sequences,padding='post')
```

```python
sequences
```

```python
sequences.shape
```

```python
Xtr,Xte,Ytr,Yte=train_test_split(sequences,Y,test_size=0.2,random_state=42)
```

```python
Xtr.shape
```

```python
Xte.shape
```

```python
from tensorflow import keras
from keras import Sequential
from keras.layers import Dense,SimpleRNN,Embedding,Flatten
```

```python
model=Sequential()
```

```python
model.add(Embedding(input_dim=len(tokenizer.word_index) + 1, output_dim=32, input_length=sequences.shape[1]))
model.add(SimpleRNN(32,input_shape=(62,1),return_sequences=False))
model.add(Dense(1,activation='sigmoid'))
model.summary()
```

```python
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
model.fit(Xtr,Ytr,epochs=1000,validation_data=(Xte,Yte))
```

```python
# Evaluate the model
loss, accuracy = model.evaluate(Xte, Yte)
print(f'Test Loss: {loss}')
print(f'Test Accuracy: {accuracy}')
```

```python
def predict_sentiment(review):
    # Preprocess the review
    review_seq = tokenizer.texts_to_sequences([review])
    review_pad = pad_sequences(review_seq, padding='post', maxlen=sequences.shape[1])
    # Predict sentiment
    prediction = model.predict(review_pad)
    return prediction[0][0]
```

```python
custom_review = "This product is amazing and I loved it"
sentiment_score = predict_sentiment(custom_review)
print(f'Sentiment score for custom review: {sentiment_score}')
```

```python
custom_review = "This product is worst ever and I just didn't liked it but, also hated it!"
sentiment_score = predict_sentiment(custom_review)
print(f'Sentiment score for custom review: {sentiment_score}')
```

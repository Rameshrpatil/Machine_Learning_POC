```python
import gensim
import numpy as np
import pandas as pd
from gensim.models import Word2Vec
from nltk.corpus import brown
from nltk.tokenize import word_tokenize
from nltk import download
```

```python
df=pd.read_csv('/content/amazon.csv')
```

```python
df
```

```python
df['reviewText']=df['reviewText'].str.lower()
```

```python
import re
def html_handler(text):
    pattern = re.compile(r'<.*?>')
    return pattern.sub(r'',text)
```

```python
df['reviewText']=df['reviewText'].apply(html_handler)
```

```python
df['reviewText']
```

```python
def  url_handler(text):
    pattern = re.compile(r'http[s]?://(?:[a-zA-Z]|[0-9]|[$-_@.&+]|[!*\\(\\),]|(?:%[0-9a-fA-F][0-9a-fA-F]))+')
    return pattern.sub(r'',text)
```

```python
df['reviewText']=df['reviewText'].apply(url_handler)
```

```python
import string
exclude = set(string.punctuation)
```

```python
def remove_punctions(text):
    for char in exclude:
        text = text.replace(char, '')
    return text
```

```python
df['reviewText']=df['reviewText'].apply(remove_punctions)
```

```python
df['reviewText']
```

```python
corpus=df['reviewText']
```

```python
corpus=np.array(corpus)
```

```python
import nltk
nltk.download('punkt')
```

```python
nltk.download('wordnet')
nltk.download('stopwords')
from nltk.stem import WordNetLemmatizer
from nltk.corpus import stopwords
```

```python
lemmatizer=WordNetLemmatizer()
```

```python
import re
sentences=[]
for i in range(len(corpus)):
    review = re.sub('[^a-zA-Z]',' ',corpus[i])
    review=review.lower()
    review=review.split()
    review=[lemmatizer.lemmatize(word) for word in review if not word in set(stopwords.words('english'))]
    review=' '.join(review)
    sentences.append(review)
```

```python
sentences
```

```python
tokenized_sentences = [word_tokenize(sentence) for sentence in sentences]
```

```python
tokenized_sentences
```

```python
model = Word2Vec(sentences=tokenized_sentences, vector_size=100, window=5, min_count=2, workers=4)
```

```python
vector = model.wv['smartphone']
print("Vector for 'smartphone':", vector)
```

```python
similar_words = model.wv.most_similar('smartphone', topn=5)
print("Words most similar to 'smartphone':", similar_words)
```
